```python
# importing the required modules

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import torch
from transformers import BertTokenizer, BertForSequenceClassification, AdamW
from torch.utils.data import Dataset, DataLoader
import torch.nn.functional as F
```

```python
#Loading the heart_disease_uci dataset
data = pd.read_csv('../data/heart_disease/heart_disease_uci.csv')

data.head()
```

```python
#Converting num to binary
data['num'] = data['num'].apply(lambda x: 1 if x>0 else 0)
data.head()
```

```python
# checking count of 1's and 0's in the target variable
data['num'].value_counts()
```

```python
# counting the number of males and females
data['sex'].value_counts()
```

```python
# Filter the dataset to include only rows where target is 1
target_1_data = data[data['num'] == 1]

# Group by the 'sex' column and count the occurrences
# Assuming 'sex' column: 1 for male and 0 for female
target_1_counts = target_1_data.groupby('sex').size()

# Display the counts
print(target_1_counts)
```
